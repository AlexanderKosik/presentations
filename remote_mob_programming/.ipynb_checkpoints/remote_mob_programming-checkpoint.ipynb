## Was ist (Remote) Mob Programming

**Mob Programming**

"Beim Mob-Programming wird Software durch mehrere Personen am selben Ort, zur selben Zeit, gemeinsam an einem Computer erstellt."

**Remote Mob Programming**

"am selben Ort" wird durch einen "virtuellen Ort" ersetzt

# 15 Essentials


## Remote Everybody

- Jeder arbeitet remote
- Keine *Informations-Asymetrie*
- Kein Pendeln, mehr Zeit fr die Famile, weniger Unterbrechungen
- ...

## Camera Always On

- Mehr Details der Kommunikation wahrnehmbar
- Weniger Nebenbeschftigungen (Smartphone, ) durch Kamera
- **Always on**: Mehr "Prsenz-Gefhl"
    - Ich kann sehen, wenn jemand AFK ist
    - Ich kann sehen, wenn jemand gerade abgelenkt wird
- Mikrofon wird beim Verlassen des Platzes gemutet, Kamera bleibt an


## Regular On-Site Meetings

- Je besser man sich kennt, desto besser kann man zusammenarbeiten
- Kennenlernen funktioniert On-Site am Besten
- 1x pro Monat: Soziales Event mit dem Team

## Small Team

- Ideale Mob-Gre: 3-4 Personen
- Je mehr Teilnehmer, desto kleiner die Sprechzeit pro Teilnehmer, mehr Strungen etc.

## Same Time

#### *Mob at least 6 hours a day*

- Zusammenarbeit: 6h am Tag
- Festlegung von "Kern-Arbeitszeit"
- Festlegung der Mittagspause
- Zeit fr sonstige "Meetings" einplanen: Besprechungen, Frisr, Kinder abholen, ...

## Typist and the Rest of the Mob

- **Typist**: Person an der Tastatur
    - Tippt 
    - Stellt Nachfragen
    - Muss die Lsung und das Problem verstehen
    
    
- **Rest of the mob**:
    - Diskutiert das Problem
    - Erstellt Lsungen
    - Instruiert den Typist
    
Dadurch:
- Balance durch wechselnde Rollen
- Ausgleich fr Dominanz bestimmter Charactere

## Screen Sharing

- Typist teilt den Bildschirm
- Diskussionsgrundlage: Code/IDE
- "Rest of the mob" navigiert durch Sprache

## 10 Minute Intervals

- Jede Mob-Session hat ein Ziel (Neues Feature, Bugfix, ...)
- Alle 10 Minuten wird gewechselt
    - Bei krzeren Zeiten: Zuviel Overhead durch Wechsel
    - Lnger? (Keine Aussage)
- Rolle des Typist ist mental entspannend (durch Navigation)

## Git Handover

- Handover erfolgt durch WIP-Commits in einem Branch
- Nach jedem Interval wird gepushed
    - Muss nicht compilieren
    - Es mssen nicht alle Tests laufen
- Schneller Handover ist entscheidend
- Tool: *mob* zur Vereinfachung des Prozess

## Group Decisions

- Ziel: Bessere Entscheidungen treffen
- Alle Entscheidungen sind Gruppenentscheidungen
    - Mehr Diskussion
    - Mehr (Erfahrungs-)Austausch

Entscheidungen werden mit *Architecture Decision Records* dokumentiert

### Durch Group Decisions:
- Kein Code Review
- Keine Pull Requests

## Constant Momentum

- Kontinuierliches, implizites Code-Review
- Kein Warten, kein Kontext-Switch
- Fortwhrender Fortschritt

## Learn from the Team

- Wissen zu teilen ist ein Kernaspekt des Mob-Programmings
- Jeder lernt beim Mob-Programming permanent etwas durch die Anderen
    - Gute Unit Tests
    - Effektives Debugging
    - Gute Meetingvorbereitungen
    - Neue Tastenkrzel
    - ...
- Bus-Factor ist die Mob-Gre
- Schnelles Onboarding


## Trust

- Durch permanente Remote-Arbeit "sieht" niemand das Team arbeiten
- Das Team muss sich das Vertrauen durch eingehaltenes Commitment und "Continuous Delivery" erarbeiten

Das Team dokumentiert tglich in einen Chat-Channel:
- Check-Ins: https://basecamp.com/features/checkins
- Kurze Zusammenfassung, was nicht funktioniert hat/wo es Probleme gab/Abweichungen vom Plan
- Sonstige Erkenntnisse

## Save the Planet

Keine Fahrten --> Keine Emissionen, keine Kosten

## Dine with your Family

Das Beste aus beiden Welten:
1. Zeit fr und mit der Famile
2. Arbeiten an herausfordernden und interessanten Projekten von zu Hause aus
